```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import os

file_path = "dummy_file.csv"
if not os.path.exists(file_path):

    dummy_data = {'col1': [1, 2, 3], 'col2': [4, 5, 6]}
    df_dummy = pd.DataFrame(dummy_data)

    df_dummy.to_csv(file_path, index=False)


df = pd.read_csv(file_path)


print(df.info())
print(df.describe())


print("\nNull values in dataset:")
print(df.isnull().sum())


print("\nSum of null values in each row:")
print(df.isnull().sum(axis=1))


df_dropna = df.dropna()


df_fillna_0 = df.fillna(0)


df_ffill = df.fillna(method='ffill')
df_bfill = df.fillna(method='bfill')


if 'col1' in df.columns:
    mean_value = df['col1'].mean()
    df_fillna_mean = df.fillna(mean_value)


df_clean = df.dropna()


age = [1,3,28,27,25,92,30,39,40,50,26,24,29,94]
af = pd.DataFrame(age, columns=['Age'])
plt.figure(figsize=(5, 3))
sns.boxplot(data=af)
plt.title("Boxplot for Age")
plt.show()


Q1 = af['Age'].quantile(0.25)
Q3 = af['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = af[(af['Age'] < lower_bound) | (af['Age'] > upper_bound)]
print("\nOutliers detected using IQR:")
print(outliers)


af_no_outliers = af[(af['Age'] >= lower_bound) & (af['Age'] <= upper_bound)]


plt.figure(figsize=(5, 3))
sns.boxplot(data=af_no_outliers)
plt.title("Boxplot After Removing Outliers")
plt.show()


data = [1,12,15,18,21,24,27,30,33,36,39,42,45,48,51,54,57,60,63,66,69,72,75,78,81,84,87,90,93,96,99,158]
df = pd.DataFrame(data, columns=['Value'])
plt.figure(figsize=(5, 3))
sns.boxplot(data=df)
plt.title("Boxplot for Data")
plt.show()

z_scores = stats.zscore(df['Value'])
outliers_z = df[abs(z_scores) > 3]
print("\nOutliers detected using Z-score:")
print(outliers_z)

df_no_outliers = df[abs(z_scores) <= 3]

plt.figure(figsize=(5, 3))
sns.boxplot(data=df_no_outliers)
plt.title("Boxplot After Removing Outliers")
plt.show()

```
